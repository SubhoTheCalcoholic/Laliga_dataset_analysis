```python
data1.replace('-',np.nan, inplace=True)
data1=data1.fillna(0)
print(data1)
```

```python
df=pd.DataFrame(data1)
```

```python
print(df.dtypes)
```

```python
df['Points'] = pd.to_numeric(df['Points'], errors='coerce')
```

```python
df['GamesWon'] = pd.to_numeric(df['GamesWon'], errors='coerce')
```

```python
df['GamesPlayed'] = pd.to_numeric(df['GamesPlayed'], errors='coerce')
```

```python
print(df.dtypes)
```

```python
data1['Debut'] = data1['Debut'].astype(str).str[:4].astype(int)
laliga_new = data1[data1['Debut'].between(1930, 1980)]
print(laliga_new)
```

```python
df.sort_values(by=['Points'],ascending = False)
```

```python
df['WinningPercent'] = (df.iloc[:,5]/df.iloc[:,4])*100
print(df)
```

```python
df['GoalsFor'] = pd.to_numeric(df['GoalsFor'], errors='coerce')
```

```python
df['GoalsAgainst'] = pd.to_numeric(df['GoalsAgainst'], errors='coerce')
```

```python
print(df.dtypes)
```

```python
df['GoalsDifferences']=df['GoalsFor']-df['GoalsAgainst']
```
